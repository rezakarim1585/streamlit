```python
import pandas as pd
import numpy as np
import sklearn
import pickle
import math

from sklearn import linear_model

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

#######
df = pd.read_csv('ssdata.csv')


df= df[['bw']+['d']+['fc']+['a_d']+['pf']+['Ef']+['Vexp']]

df=pd.DataFrame(df)

X = df.loc[:, df.columns != 'Vexp']


df['bw'] = np.log(df['bw'])
df['d'] = np.log(df['d'])
df['fc'] = np.log(df['fc'])
df['a_d'] = np.log(df['a_d'])
#df['pf'] = np.log(df['pf'])
df['Ef'] = np.log(df['Ef'])
#df['Vexp'] = np.log(df['Vexp'])

df=pd.DataFrame(df)
#st.write(df)

X1 = df.loc[:, df.columns != 'Vexp']
Y=df['Vexp']

########
from sklearn.linear_model import LinearRegression

lr_model = linear_model.LinearRegression()
lr_model.fit(X1, Y)

model_lr = "lr_model.pkl"
with open(model_lr,'wb') as f:
    pickle.dump(lr_model, f)
###########
from sklearn.linear_model import Ridge

rr_model = linear_model.Ridge(alpha=0.001, random_state=3)
rr_model.fit(X1, Y)

model_rr = "rr_model.pkl"
with open(model_rr,'wb') as f:
    pickle.dump(rr_model, f)
############
from sklearn.linear_model import Lasso

lasr_model = linear_model.Lasso(alpha=0.001)
lasr_model.fit(X1, Y)

model_lasr = "lasr_model.pkl"
with open(model_lasr,'wb') as f:
    pickle.dump(lasr_model, f)
########

from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X1, Y)

model_knn = "knn_model.pkl"
with open(model_knn,'wb') as f:
    pickle.dump(knn_model, f)

########
from sklearn.svm import SVR

svr_model = SVR(kernel='rbf', C=21, degree=3)
svr_model.fit(X1, Y)

model_svr = "svr_model.pkl"
with open(model_svr,'wb') as f:
    pickle.dump(svr_model, f)

##########
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=3)
dt_model.fit(X1, Y)

model_file = "dt_model.pkl"
with open(model_file,'wb') as f:
    pickle.dump(dt_model, f)    
############    
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()
rf_model.fit(X1, Y)

model_file = "rf_model.pkl"
with open(model_file,'wb') as f:
    pickle.dump(rf_model, f)

##############    
import xgboost as xgb
from xgboost import XGBRegressor

xb_model = xgb.XGBRegressor(n_estimators=200, learning_rate=0.16, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)
xb_model.fit(X1, Y)

model_filexb = "xb_model.pkl"
with open(model_filexb,'wb') as f:
    pickle.dump(xb_model, f)
########

from sklearn.ensemble import AdaBoostRegressor

ab_model= AdaBoostRegressor(DecisionTreeRegressor(max_depth=9),random_state=3,n_estimators=750, learning_rate=0.16)
                           
ab_model.fit(X1, Y)

model_ab = "ab_model.pkl"
with open(model_ab,'wb') as f:
    pickle.dump(ab_model, f)

##############
from sklearn.neural_network import MLPRegressor
ann_model = MLPRegressor(hidden_layer_sizes=(128,128,128,128,128),activation="relu",random_state=3, max_iter=1500).fit(X1, Y)

model_ann = "ann_model.pkl"
with open(model_ann,'wb') as f:
    pickle.dump(ann_model, f)
    
##########

from catboost import CatBoostRegressor
cb_model = CatBoostRegressor(iterations=700,learning_rate=0.2,depth=16,eval_metric='RMSE',random_seed = 23,bagging_temperature = 0.2,od_type='Iter',
                             metric_period = 75,
                             od_wait=100)

cb_model.fit(X1, Y)    

model_cb = "cb_model.pkl"
with open(model_cb,'wb') as f:
    pickle.dump(cb_model, f)
    
    


```
